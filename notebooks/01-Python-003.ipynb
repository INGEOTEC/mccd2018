```python
'''Agregar 10 registros, uno por cada compañero de clase 
(puede inventar los datos, excepto el nombre)
'''
import random
import numpy as np
d_temp=DataFrame(columns=n_datos.columns)

Alumnos=["Manuel","Santiago","Hugo","Briana","Olivia","Axel","Rodrigo",
         "Daniel","Dulce","Hector"]

for i in range(0,10):
    d_temp.loc[i,"COMUNA"] = random.choice(["Arauco",np.nan])
    d_temp.loc[i,"LOCALIDAD"]= random.choice(["Arauco","Carampangue",
                                              "Laraquete",np.nan])
    d_temp.loc[i,"TIPO DE BIEN"]= random.choice(["departamento",
                                                 "Casa",np.nan])
    d_temp.loc[i,"NORMATIVA"]= random.choice(["Urbana",
                                              "Rural",np.nan])
    d_temp.loc[i,"PROPIETARIO"]= Alumnos[i]
    d_temp.loc[i,"Edad"]= random.randint(20,25)
    
d_temp
```

## Ejemplo Titanic 
- Primero debemos cargar los datos de training.


```python
train = pd.read_csv("train.csv",index_col = 'PassengerId') 
train = pd.DataFrame(train)
```

```python
train
```

```python
train.describe() # Nos resume información estadística global del dataset
```

```python
print(train.columns) #Nos lista el nombre de las columnas
print(train.shape) # nos indica el número de filas y columnas
```

```python
## Estructurando los datos.
# Se llenan los missing values de edad.
train['Age'].fillna(train['Age'].median(), inplace=True)

# ¿Para qué nos sirve el nombre?
train['Mr'] = 0
train['Mrs'] = 0
train['Miss'] = 0
train['royalty'] = 0
train['officer'] = 0
```

```python
#Se categoriza, en relación con las columnas anteriores,
# el nombre del pasajero
for index, row in train.iterrows():
    name = row['Name']
    if 'Mr.' in name:
        train.set_value(index,'Mr',1)
    elif 'Mrs.' in name:
        train.set_value(index,'Mrs',1)
    elif 'Miss.' in name:
        train.set_value(index,'Miss',1)
    elif 'Lady' or 'Don' or 'Dona' or 'sir' or 'master' in name:
        train.set_value(index,'royalty',1)
    elif 'rev' in name:
        train.set_value(index,'officer',1)
        
train.head(2)
train.tail(2)
```

```python
train.drop('Name',inplace=True, axis=1)
```

```python
train.head()
```

```python
#Port of Embarkation: (C = Cherbourg; Q = Queenstown; S = Southampton)
train['Embarked_S'] = 0
train['Embarked_C'] = 0
train['Embarked_Q'] = 0
train['Embarked_unknown'] = 0

for index,row in train.iterrows():
    embarkment = row['Embarked']
    if embarkment == 'S':
        train.set_value(index,'Embarked_S',1)
    elif embarkment == 'C':
        train.set_value(index,'Embarked_C',1)
    elif embarkment == 'Q':
        train.set_value(index,'Embarked_Q',1)
    else:
        train.set_value(index,'Embarked_unknown',1)

train.head()
```

```python
#Borramos la columna Embarked
train.drop('Embarked', inplace = True, axis = 1) 
train.head()
```

```python
# Sex 1-0
for index,row in train.iterrows():
    if row['Sex'] == 'male':
        train.set_value(index, 'Sex', 1)
    else:
        train.set_value(index,'Sex',0)
train.head()
```

```python
#Como no se utilizará la columna Ticket, lo borramos

train.drop('Ticket', inplace= True, axis = 1)
train.head()
```

```python
#Categorización de precios

train['Fare_cheap']=0
train['Fare_average']=0
train['Fare_costly']=0

for index,row in train.iterrows():
    if row['Fare'] <= 30.0 :
        train.set_value(index, 'Fare_cheap', 1)
    elif row['Fare'] >30 and  row['Fare'] <= 70.0:
        train.set_value(index,'Fare_average',1)
    else:
        train.set_value(index, 'Fare_costly',1)
        
train.head()
```

```python
train.drop('Fare',inplace = True, axis =1) # Borramos la columna
train.head()
```

```python
#Como tampoco se usará la columna Cabin, la eliminamos.
train.drop('Cabin',inplace = True, axis = 1)
train.head()
```

```python
train.describe() # Se vuelve a verificar si no hay algún valor faltante después de las modificaciones
```

```python
#Dividimos datos para entrenamiento
X = train[['Pclass','Sex','Age','SibSp','Parch','Mr','Mrs','Miss',
           'royalty','officer',
           'Embarked_S','Embarked_C','Embarked_Q',
           'Embarked_unknown','Fare_cheap',
           'Fare_average','Fare_costly']]
X
```

```python
#Guardamos en Y las etiquetas de la clase: sobrevivió: 1 ó 0.
y = train.Survived
y
```
